```python
nhldraft_data = nhldraft_data.drop_duplicates(subset=['player', 'nationality', 'amateur_team', 'points'], keep='first')

nhldraft_data.value_counts('player').head(20)
```

```python
craig_anderson = nhldraft_data[nhldraft_data['player'] == 'Craig Anderson']
craig_anderson
```

```python
nhldraft_data.dropna(subset = ['team'], inplace = True)
```

```python
# Replace NaN with 0
nhldraft_data = nhldraft_data.fillna(0)
```

```python
# Most draft picks in the top 10 grouped by team
most_top10_picks = nhldraft_data[(nhldraft_data['overall_pick'] > 1) & (nhldraft_data['overall_pick'] <= 10)].groupby('team', as_index = False)['id'].count()
most_top10_picks.sort_values('id', ascending = False)
```

```python
# What team has had the most picks ever?
most_picks = nhldraft_data.groupby('team', as_index = False)['id'].count()
most_picks.sort_values('id', ascending = False)
```

```python
# What team has drafted the players with the most games played on avg?
most_games_played = nhldraft_data.groupby('team', as_index = False)['games_played'].mean()
most_games_played.sort_values('games_played', ascending = False)
```

```python
most_games_played_by_whaler = nhldraft_data[nhldraft_data['team'] == 'Hartford Whalers'][['player', 'team', 'games_played']]
most_games_played_by_whaler.sort_values('games_played', ascending = False)
```

```python
# What team has drafted the players with the most goals scored on avg
most_goals_scored_team = nhldraft_data.groupby('team', as_index = False)['goals'].mean()
most_goals_scored_team.sort_values('goals', ascending = False)
```

```python
most_goals_scored_by_scouts = nhldraft_data[nhldraft_data['team'] == 'Kansas City Scouts'][['player', 'team', 'goals']]
most_goals_scored_by_scouts.sort_values('goals', ascending = False)
```

```python
# What team has drafted the most players that have played atleast one game?
most_drafted_players = nhldraft_data[nhldraft_data['games_played'] > 0].groupby('team', as_index = False).count()[['team', 'id']]
most_drafted_players.sort_values('id', ascending = False)
```

```python
# Which team has the best success rate of players drafted playing atleast 1 game in the NHL?

# calculate the total number of players drafted by each team
picks_00_to_20 = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2020 )].groupby('team', as_index = False).count()[['team', 'id']]
# calculate the number of players drafted to actually play atleast 1 game
successful_picks_00_to_20 = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2020) & (nhldraft_data['games_played'] > 0)].groupby('team', as_index = False).count()[['team', 'id']]
# rename columns
picks_00_to_20 = picks_00_to_20.rename(columns={"id": "total_picks"}, errors = "raise")
successful_picks_00_to_20 = successful_picks_00_to_20.rename(columns = {'id': "successful_picks"}, errors = 'raise')

# combine data frames
frames = [picks_00_to_20, successful_picks_00_to_20]
draft_success = pd.concat(frames)
draft_success = draft_success.groupby('team', as_index = False).sum()
# create new column that holds the success rate
draft_success['success_rate'] = (draft_success['successful_picks'] / draft_success['total_picks']) * 100
draft_success.sort_values('success_rate', ascending = False)
```

```python
# Calculate the number of teams per year
unique_teams = nhldraft_data.groupby('year', as_index = False)['team'].nunique()
unique_teams
```

```python
plt.figure(figsize = (10.5, 7.25))
fig = plt.figure()

ax = unique_teams.plot(x = 'year', y = 'team', legend = None)
ax.set_title('Number of Teams Participating in the Draft by Year', fontsize = 16)
ax.set(xlabel = 'Year', ylabel = 'Number of Teams')
plt.show()

# ax = round_success.plot.bar(x = 'round', stacked = True, color = ['lightseagreen', 'tomato'], figsize = (8, 6))
# ax.set_title('Success Rate of Draft Picks by Round', fontsize = 16)
# ax.set_ylim(0, 550)
# ax.set_xticklabels(['1', '2', '3', '4', '5', '6', '7'], rotation = 0)
# ax.set(xlabel="Round", ylabel="Players Drafted")
```

```python
# Success rate per round of draft from 2000-16
# only include years 2000 to 2016 because those are the years that had 30 teams
#

def make_rounds(overall_pick):
    if overall_pick <= 30:
        return 1
    elif overall_pick <= 60:
        return 2
    elif overall_pick <= 90:
        return 3
    elif overall_pick <= 120:
        return 4
    elif overall_pick <= 150:
        return 5
    elif overall_pick <= 180:
        return 6
    elif overall_pick <= 210:
        return 7
#     elif overall_pick <= 240:
#         return 8
#     elif overall_pick <= 270:
#         return 9
#     elif overall_pick <= 300:
#         return 10
    
nhldraft_data['round'] = nhldraft_data['overall_pick'].apply(make_rounds)
per_round = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2016)]

picks_per_round = per_round.groupby('round', as_index = False).count()[['round', 'id']]
picks_per_round = picks_per_round.rename(columns = {'id': 'picks'}, errors = "raise")

success_per_round = per_round[per_round['games_played'] > 0].groupby('round', as_index = False).count()[['round', 'id']]
success_per_round = success_per_round.rename(columns={"id": "successful_picks"}, errors = "raise")



frames = [picks_per_round, success_per_round]
round_success = pd.concat(frames)
round_success = round_success.groupby('round', as_index = False).sum()
round_success['unsuccessful'] = (round_success['picks'] - round_success['successful_picks'])
round_success = round_success[['round', 'successful_picks', 'unsuccessful']]
round_success
```

```python
plt.figure(figsize = (10.5, 7.25))
fig = plt.figure()

ax = round_success.plot.bar(x = 'round', stacked = True, color = ['lightseagreen', 'tomato'], figsize = (8, 6))
ax.set_title('Success Rate of Draft Picks by Round', fontsize = 16)
ax.set_ylim(0, 550)
ax.set_xticklabels(['1', '2', '3', '4', '5', '6', '7'], rotation = 0)
ax.set(xlabel="Round", ylabel="Players Drafted")


# ax = df.plot.bar(x='School', stacked=True, color=['tomato','lightseagreen'], figsize=(8,6))
# ax.set_title('The Number of Students', fontsize=20)
# ax.set_ylim(0,500)
# ax.set_xticklabels(['A','B','C'], rotation=0)
```

```python
# Nationalities
nationalities = nhldraft_data.groupby('nationality', as_index = False)['id'].count()
nationalities = nationalities[nationalities['id'] > 100]
nationalities
```

```python
x = nationalities['id']
explode = (0, 0.1, 0, 0, 0, 0.1, 0)
colors = ['salmon', 'cyan', 'violet', 'palegreen', 'dodgerblue', 'gold', 'turquoise']
autopct = '%1.1f%%'

plt.figure(figsize = [15, 15])
fig, ax = plt.subplots()
ax.pie(x,
      explode = explode,
      labels = ['Canada', 'Czech', 'Finland', 'Russia', 'Sweden', 'Slovakia', 'United States'],
      colors = colors,
      autopct = autopct,
      shadow = True,
      radius = 2,
      startangle = 90)
ax.axis('equal')

plt.title("Nationalities in the NHL Draft 1963 - Present")
plt.show()
```

```python
nationalities_22 = nhldraft_data[(nhldraft_data['year'] >= 2018) & (nhldraft_data['year'] <= 2022)].groupby('nationality', as_index = False)['id'].count()
nationalities_22 = nationalities_22[nationalities_22['id'] > 15]
nationalities_22
```

```python
x = nationalities_22['id']
explode = (0, 0, 0.1, 0, 0, 0, 0)
colors = ['salmon', 'cyan', 'violet', 'palegreen', 'dodgerblue', 'gold', 'turquoise']
autopct = '%1.1f%%'

plt.figure(figsize = [15, 15])
fig, ax = plt.subplots()
ax.pie(x,
       explode = explode,
       labels = ['Canada', 'Czech', 'Denmark', 'Finland', 'Russia', 'Sweden', 'United States'],
       colors = colors,
       autopct = autopct,
       shadow = True,
       radius = 2,
       startangle = 90)
ax.axis('equal')

plt.title("Nationalities in the NHL Draft 2018 - Present")
plt.show()
```
