```python
# Replace NaN with 0
nhldraft_data = nhldraft_data.fillna(0)

nhldraft_data['games_played'] = nhldraft_data['games_played'].apply(int)
```

```python
# Most draft picks in the top 10 grouped by team
most_top10_picks = nhldraft_data[(nhldraft_data['overall_pick'] > 1) & (nhldraft_data['overall_pick'] <= 10)].groupby('team', as_index = False)['id'].count()
most_top10_picks.sort_values('id', ascending = False)
```

```python
# What team has had the most picks ever?
most_picks = nhldraft_data.groupby('team', as_index = False)['id'].count()
most_picks.sort_values('id', ascending = False).head(5)
```

```python
# What team has drafted the players with the most games played on avg?
most_games_played = nhldraft_data.groupby('team', as_index = False)['games_played'].mean()
most_games_played.sort_values('games_played', ascending = False).head(5)
```

```python
# Most drafted players with over 1000 games
most_1000_gp = nhldraft_data[nhldraft_data['games_played'] > 1000].groupby('team', as_index = False)['games_played'].count()
most_1000_gp.sort_values('games_played', ascending = False).head(5)
```

```python
# What team has drafted the most players that have played atleast one game?
most_drafted_players = nhldraft_data[nhldraft_data['games_played'] > 0].groupby('team', as_index = False).count()[['team', 'id']]
most_drafted_players.sort_values('id', ascending = False)
```

```python
# Which team has the best success rate of players drafted playing atleast 82 games in the NHL from 2000-2020?

# calculate the total number of players drafted by each team
picks_00_to_20 = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2020 )].groupby('team', as_index = False).count()[['team', 'id']]
# calculate the number of players drafted to actually play atleast 1 game
successful_picks_00_to_20 = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2020) & (nhldraft_data['games_played'] > 82)].groupby('team', as_index = False).count()[['team', 'id']]
# rename columns
picks_00_to_20 = picks_00_to_20.rename(columns={"id": "total_picks"}, errors = "raise")
successful_picks_00_to_20 = successful_picks_00_to_20.rename(columns = {'id': "successful_picks"}, errors = 'raise')

# combine data frames
frames = [picks_00_to_20, successful_picks_00_to_20]
draft_success = pd.concat(frames)
draft_success = draft_success.groupby('team', as_index = False).sum()
# create new column that holds the success rate
draft_success['success_rate'] = (draft_success['successful_picks'] / draft_success['total_picks']) * 100
draft_success.sort_values('success_rate', ascending = False).head()
```

```python
plt.figure(figsize = (10.5, 7.25))
fig = plt.figure()

x = draft_success['team']
height = draft_success['success_rate']

plt.bar(x,
      height,
      color = ['lightcoral', 'lightskyblue', 'lightgreen', 'plum', 'moccasin'],
      width = 0.5)
#plt.xticks(np.arange(5), ('Center', 'Defense', 'Goalie', 'Left Wing', 'Right Wing'), rotation = 45)
plt.xticks(rotation = 90)
plt.xlabel('Team')
plt.ylabel('Percentage of Successful Picks')
plt.title('Percentage of Draft Picks to Play atleast 82 Games')
plt.ylim(0, 100)
plt.grid()

plt.show()
```

```python
# Success rate per round of draft from 2000-16
# only include years 2000 to 2016 because those are the years that had 30 teams
#

def make_rounds(overall_pick):
    if overall_pick <= 30:
        return 1
    elif overall_pick <= 60:
        return 2
    elif overall_pick <= 90:
        return 3
    elif overall_pick <= 120:
        return 4
    elif overall_pick <= 150:
        return 5
    elif overall_pick <= 180:
        return 6
    elif overall_pick <= 210:
        return 7

# Create a new column called 'round' that holds a value showing which round of the draft each player was taken
nhldraft_data['round'] = nhldraft_data['overall_pick'].apply(make_rounds)

# Filter our data down to the years from 2000-2016
per_round = nhldraft_data[(nhldraft_data['year'] >= 2000) & (nhldraft_data['year'] <= 2016)]

picks_per_round = per_round.groupby('round', as_index = False).count()[['round', 'id']]
picks_per_round = picks_per_round.rename(columns = {'id': 'picks'}, errors = "raise")

success_per_round = per_round[per_round['games_played'] > 82].groupby('round', as_index = False).count()[['round', 'id']]
success_per_round = success_per_round.rename(columns={"id": "successful_picks"}, errors = "raise")



frames = [picks_per_round, success_per_round]
round_success = pd.concat(frames)
round_success = round_success.groupby('round', as_index = False).sum()
round_success['unsuccessful'] = (round_success['picks'] - round_success['successful_picks'])
round_success = round_success[['round', 'successful_picks', 'unsuccessful']]
round_success.to_csv('round_success.csv')
round_success
```

```python
fig = plt.figure()

ax = round_success.plot.bar(x = 'round', stacked = True, color = ['lightseagreen', 'tomato'], figsize = (8, 6))
ax.set_title('Success Rate of Draft Picks by Round (2000-2016)', fontsize = 16)
ax.set_ylim(0, 550)
ax.set_xticklabels(['1', '2', '3', '4', '5', '6', '7'], rotation = 0)
ax.set(xlabel="Round", ylabel="Players Drafted")
fig.savefig('round_success')

plt.show()
```

```python
# Nationalities
nationalities = nhldraft_data.groupby('nationality', as_index = False)['id'].count()
nationalities = nationalities[nationalities['id'] > 100]
nationalities
```

```python
x = nationalities['id']
explode = (0, 0.1, 0, 0, 0, 0.1, 0)
colors = ['salmon', 'cyan', 'violet', 'palegreen', 'dodgerblue', 'gold', 'turquoise']
autopct = '%1.1f%%'

plt.figure(figsize = [15, 15])
fig, ax = plt.subplots()
ax.pie(x,
      explode = explode,
      labels = ['Canada', 'Czech', 'Finland', 'Russia', 'Sweden', 'Slovakia', 'United States'],
      colors = colors,
      autopct = autopct,
      shadow = True,
      radius = 2,
      startangle = 90)
ax.axis('equal')

plt.title("Nationalities in the NHL Draft (1963-2022)")
fig.savefig('nationalities_1963')
plt.show()
```
